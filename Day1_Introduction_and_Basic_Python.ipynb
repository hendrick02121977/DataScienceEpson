## Part 2: Python Fundamentals

### 2.1 Variables and Data Types

```python
# --- Variables and Basic Data Types ---

# Integer
machine_id = 101
print(f"Machine ID: {machine_id} (type: {type(machine_id).__name__})")

# Float
temperature = 72.5          # degrees Celsius
pressure = 1013.25          # hPa
print(f"Temperature: {temperature} °C, Pressure: {pressure} hPa")

# String
machine_name = "Epson_CNC_Line_A"
status = "RUNNING"
print(f"Machine: {machine_name}, Status: {status}")

# Boolean
is_operational = True
alert_triggered = False
print(f"Operational: {is_operational}, Alert: {alert_triggered}")
```

```python
# --- Collections ---

# List: ordered, mutable
sensor_readings = [72.1, 72.5, 73.0, 74.2, 71.8, 75.0, 70.5]
print(f"Sensor readings: {sensor_readings}")
print(f"First reading: {sensor_readings[0]}, Last: {sensor_readings[-1]}")
print(f"Slice [2:5]: {sensor_readings[2:5]}")

# Tuple: ordered, immutable
coordinates = (7.5, 23.1)   # (x, y) position on factory floor
print(f"\nMachine coordinates: {coordinates}")

# Dictionary: key-value pairs
machine_info = {
    "id": 101,
    "name": "Epson_CNC_Line_A",
    "temperature": 72.5,
    "status": "RUNNING",
    "cycles_today": 1250
}
print(f"\nMachine info: {machine_info}")
print(f"Cycles today: {machine_info['cycles_today']}")

# Set: unordered, unique elements
error_codes = {"E001", "E003", "E001", "E007", "E003"}
print(f"\nUnique error codes: {error_codes}")
```

### 2.2 Flow Control

```python
# --- Conditional Statements ---

def check_temperature(temp, normal_min=60, normal_max=80, critical_max=95):
    """Classify machine temperature into operational zones."""
    if temp > critical_max:
        return "CRITICAL — STOP MACHINE"
    elif temp > normal_max:
        return "WARNING — Above normal range"
    elif temp < normal_min:
        return "WARNING — Below normal range"
    else:
        return "OK — Normal operation"

test_temps = [55, 72.5, 83, 97]
for t in test_temps:
    print(f"Temperature {t}°C: {check_temperature(t)}")
```

```python
# --- Loops ---

# for loop: iterate over sensor readings
print("=== Sensor Reading Analysis ===")
total = 0
above_threshold = []
threshold = 73.0

for i, reading in enumerate(sensor_readings):
    total += reading
    status = "HIGH" if reading > threshold else "normal"
    print(f"  Reading {i+1}: {reading:.1f}°C [{status}]")
    if reading > threshold:
        above_threshold.append(reading)

average = total / len(sensor_readings)
print(f"\nAverage: {average:.2f}°C")
print(f"Readings above {threshold}°C: {above_threshold}")
```

```python
# --- While loop and List Comprehension ---

# while loop: simulate monitoring until alert
readings_stream = [70.1, 71.2, 74.5, 78.3, 82.1, 90.5, 96.0]
alarm_threshold = 85.0
idx = 0
print("Monitoring machine temperature...")
while idx < len(readings_stream):
    r = readings_stream[idx]
    print(f"  t={idx}: {r}°C")
    if r > alarm_threshold:
        print(f"  ⚠️  ALARM triggered at t={idx}: {r}°C > {alarm_threshold}°C")
        break
    idx += 1

# List comprehension: Celsius to Fahrenheit
fahrenheit = [round(c * 9/5 + 32, 1) for c in sensor_readings]
print(f"\nSensor readings in Fahrenheit: {fahrenheit}")
```

```python
# --- Functions ---

def calculate_oee(availability, performance, quality):
    """
    Calculate Overall Equipment Effectiveness (OEE).
    OEE = Availability × Performance × Quality
    All inputs should be in percentage (0-100).
    """
    oee = (availability / 100) * (performance / 100) * (quality / 100) * 100
    return round(oee, 2)

# Example: Machine Line A
availability = 90.5   # % of planned production time machine was available
performance = 95.0    # % of theoretical maximum speed
quality = 98.2        # % of good products vs total produced

oee = calculate_oee(availability, performance, quality)
print(f"Machine Line A OEE: {oee}%")
print(f"  Availability: {availability}%")
print(f"  Performance:  {performance}%")
print(f"  Quality:      {quality}%")

# World-class OEE benchmark is 85%
benchmark = 85.0
if oee >= benchmark:
    print(f"  ✅ OEE {oee}% meets world-class benchmark ({benchmark}%)")
else:
    print(f"  ⚠️  OEE {oee}% is below world-class benchmark ({benchmark}%)")
```

## Part 3: NumPy for Numerical Computing

```python
import numpy as np

# Create arrays from sensor data
np.random.seed(42)
n_samples = 100

# Simulate temperature sensor (°C) — normally distributed around 72°C
temp_data = np.random.normal(loc=72.0, scale=3.0, size=n_samples)

# Simulate vibration sensor (mm/s) — slight positive skew
vibration_data = np.random.exponential(scale=2.5, size=n_samples) + 0.5

# Simulate production count per hour
production_count = np.random.randint(low=180, high=220, size=n_samples)

print("=== NumPy Array Statistics ===")
print(f"\nTemperature Sensor (°C):")
print(f"  Shape: {temp_data.shape}, Dtype: {temp_data.dtype}")
print(f"  Min: {temp_data.min():.2f}, Max: {temp_data.max():.2f}")
print(f"  Mean: {temp_data.mean():.2f}, Std: {temp_data.std():.2f}")

print(f"\nVibration Sensor (mm/s):")
print(f"  Mean: {vibration_data.mean():.2f}, Median: {np.median(vibration_data):.2f}")

print(f"\nProduction Count (units/hour):")
print(f"  Total units: {production_count.sum()}")
print(f"  Average/hour: {production_count.mean():.1f}")
```

```python
# --- NumPy Array Operations ---

# Boolean masking: find overheating events
overheat_mask = temp_data > 78.0
overheat_events = temp_data[overheat_mask]
print(f"Overheating events (>78°C): {len(overheat_events)} out of {n_samples}")
print(f"Overheat values: {np.round(overheat_events, 1)}")

# Normalization (Min-Max scaling)
temp_normalized = (temp_data - temp_data.min()) / (temp_data.max() - temp_data.min())
print(f"\nNormalized temperature range: [{temp_normalized.min():.2f}, {temp_normalized.max():.2f}]")

# Stacking arrays into a 2D matrix
sensor_matrix = np.column_stack([temp_data, vibration_data, production_count])
print(f"\nSensor matrix shape: {sensor_matrix.shape}")
print(f"First 5 rows (temp, vibration, production):")
print(np.round(sensor_matrix[:5], 2))
```

## Part 4: Pandas for Data Analysis

```python
import pandas as pd
from datetime import datetime, timedelta

# Create a realistic production sensor DataFrame
np.random.seed(42)
n = 200

# Generate timestamps (hourly for ~8 days)
start_time = datetime(2024, 1, 1, 6, 0, 0)
timestamps = [start_time + timedelta(hours=i) for i in range(n)]

machines = ["Line_A", "Line_B", "Line_C", "Line_D"]

df = pd.DataFrame({
    "timestamp": timestamps,
    "machine_id": np.random.choice(machines, size=n),
    "temperature_c": np.round(np.random.normal(72, 3, n), 2),
    "vibration_mms": np.round(np.random.exponential(2.5, n) + 0.5, 3),
    "pressure_hpa": np.round(np.random.normal(1013, 5, n), 1),
    "production_units": np.random.randint(180, 220, n),
    "defect_count": np.random.randint(0, 8, n),
})

# Add derived columns
df["defect_rate"] = np.round(df["defect_count"] / df["production_units"] * 100, 2)
df["shift"] = df["timestamp"].apply(lambda t: "Day" if 6 <= t.hour < 18 else "Night")

print("=== Production Sensor DataFrame ===")
print(f"Shape: {df.shape}")
df.head(10)
```

```python
# --- DataFrame Inspection ---
print("Data Types:")
print(df.dtypes)
print("\nBasic Statistics:")
df.describe()
```

```python
# --- Filtering & Selection ---

# Select specific columns
temp_pressure = df[["timestamp", "machine_id", "temperature_c", "pressure_hpa"]]
print("Temperature and Pressure (first 5 rows):")
print(temp_pressure.head())

# Filter: high temperature events
high_temp = df[df["temperature_c"] > 77]
print(f"\nHigh temperature events (>77°C): {len(high_temp)}")

# Filter: Line_A with high defect rate
line_a_issues = df[(df["machine_id"] == "Line_A") & (df["defect_rate"] > 2.5)]
print(f"Line_A high defect rate (>2.5%) events: {len(line_a_issues)}")
```

```python
# --- GroupBy Aggregation ---

# Average metrics per machine
machine_summary = df.groupby("machine_id").agg(
    avg_temperature=("temperature_c", "mean"),
    avg_vibration=("vibration_mms", "mean"),
    total_production=("production_units", "sum"),
    total_defects=("defect_count", "sum"),
    avg_defect_rate=("defect_rate", "mean")
).round(2)

print("=== Machine Performance Summary ===")
print(machine_summary)

# Shift comparison
shift_summary = df.groupby("shift").agg(
    avg_temperature=("temperature_c", "mean"),
    total_production=("production_units", "sum"),
    avg_defect_rate=("defect_rate", "mean")
).round(2)

print("\n=== Day vs Night Shift Summary ===")
print(shift_summary)
```

## Part 5: Case Study — Simple Analysis of Production Sensor Data

```python
import matplotlib.pyplot as plt

# Simple 4-panel analysis plot
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle("Epson Production Sensor Data — Day 1 Case Study", fontsize=14, fontweight="bold")

# 1. Temperature over time (first 48 hours)
df_48h = df.head(48)
axes[0, 0].plot(df_48h["timestamp"], df_48h["temperature_c"], color="tomato", linewidth=1)
axes[0, 0].axhline(y=78, color="red", linestyle="--", linewidth=0.8, label="Warning threshold")
axes[0, 0].set_title("Temperature Over Time (48h)")
axes[0, 0].set_ylabel("Temperature (°C)")
axes[0, 0].tick_params(axis="x", rotation=30)
axes[0, 0].legend()

# 2. Production units per machine (bar chart)
prod_by_machine = df.groupby("machine_id")["production_units"].sum()
axes[0, 1].bar(prod_by_machine.index, prod_by_machine.values, color=["#4C72B0","#55A868","#C44E52","#8172B2"])
axes[0, 1].set_title("Total Production by Machine")
axes[0, 1].set_ylabel("Total Units")
axes[0, 1].set_xlabel("Machine")

# 3. Temperature histogram
axes[1, 0].hist(df["temperature_c"], bins=20, color="steelblue", edgecolor="white")
axes[1, 0].set_title("Temperature Distribution")
axes[1, 0].set_xlabel("Temperature (°C)")
axes[1, 0].set_ylabel("Frequency")

# 4. Defect rate by machine (box plot)
machine_order = ["Line_A", "Line_B", "Line_C", "Line_D"]
data_for_box = [df[df["machine_id"] == m]["defect_rate"].values for m in machine_order]
axes[1, 1].boxplot(data_for_box, labels=machine_order, patch_artist=True)
axes[1, 1].set_title("Defect Rate Distribution by Machine")
axes[1, 1].set_ylabel("Defect Rate (%)")

plt.tight_layout()
plt.savefig("day1_sensor_analysis.png", dpi=100, bbox_inches="tight")
plt.show()
print("\n✅ Chart saved as day1_sensor_analysis.png")
```

```python
# --- Summary Findings ---
print("=" * 55)
print("  CASE STUDY SUMMARY: Epson Production Sensor Data")
print("=" * 55)

best_machine = machine_summary["avg_defect_rate"].idxmin()
worst_machine = machine_summary["avg_defect_rate"].idxmax()
highest_prod = machine_summary["total_production"].idxmax()
overheating_pct = (df["temperature_c"] > 78).mean() * 100

print(f"\n📊 Dataset: {len(df)} hourly records across {df['machine_id'].nunique()} machines")
print(f"📅 Period: {df['timestamp'].min().date()} to {df['timestamp'].max().date()}")
print(f"\n🏆 Best quality machine: {best_machine}")
print(f"   Avg defect rate: {machine_summary.loc[best_machine, 'avg_defect_rate']:.2f}%")
print(f"\n⚠️  Worst quality machine: {worst_machine}")
print(f"   Avg defect rate: {machine_summary.loc[worst_machine, 'avg_defect_rate']:.2f}%")
print(f"\n🔥 Overheating events (>78°C): {overheating_pct:.1f}% of all readings")
print(f"\n🏭 Highest production: {highest_prod} ({machine_summary.loc[highest_prod, 'total_production']:,} units)")

print("\n📝 Key Takeaways:")
print("  1. Temperature distribution is approximately normal — expected for stable processes")
print("  2. Defect rates vary across machines — investigate root causes")
print("  3. Overheating events should trigger maintenance inspection")
```
